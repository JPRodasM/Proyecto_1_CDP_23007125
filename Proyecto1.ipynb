## Funcion para entrenar al modelo

#### Modelo Manual

```python
def training_LM(x_dep,y_ind,epoch,imprimir_error_cada,alpha):

    x_hat=[x_dep,np.ones(np.size(x_dep))]

    beta_0=0.01 #0
    beta_1=0.02 #75000
    beta=np.array([beta_1,beta_0])
    beta_dic={}

    y_hat=np.dot(np.transpose(x_hat),beta)
    e=[]

    print_flag=1

    for i in range(0,epoch):
        y_diff=y_ind-y_hat
        y_diffsq=y_diff**2

        error=(1/(2*np.size(y_hat)))*np.dot(y_diffsq,np.ones(np.size(y_diffsq)))
        e.append(error)
        
        if imprimir_error_cada!=0:
            if (print_flag<imprimir_error_cada):
                print_flag+=1
            else:
                print(error)
                print_flag=1

        de_db1=(1/(np.size(y_hat)))*np.dot(((y_hat-y_ind)*x_dep),np.ones(np.size(y_diff)))
        de_db0=(1/(np.size(y_hat)))*np.dot((y_hat-y_ind),np.ones(np.size(y_diff)))

        beta_0=beta_0-alpha*de_db0
        beta_1=beta_1-alpha*de_db1
        beta=np.array([beta_1,beta_0])
        beta_dic.update({i:beta})

        y_hat=np.dot(np.transpose(x_hat),beta)
    
    return e, beta_dic
```

## Creando Clase

```python
class RegresionLineal():

    # CONSTRUCTOR =================================================================================
    def __init__(self,X_Training,Y_Training,alpha,epoch):
        self.Training_Data_X = X_Training
        self.Training_Data_Y = Y_Training
        self.Alpha = alpha
        self.Epoch = epoch
        self.Betas = {}
        self.Errores = []
    # =============================================================================================

    # FUNCION ENTRENAMIENTO MANUAL ================================================================
    def training_LM(self,imprimir_error_cada):

        x_hat=[self.Training_Data_X,np.ones(np.size(self.Training_Data_X))]

        beta_0=0.01 
        beta_1=0.02 
        beta=np.array([beta_1,beta_0])

        y_hat=np.dot(np.transpose(x_hat),beta)

        print_flag=1

        for i in range(0,self.Epoch):
            y_diff=self.Training_Data_Y-y_hat
            y_diffsq=y_diff**2

            error=(1/(2*np.size(y_hat)))*np.dot(y_diffsq,np.ones(np.size(y_diffsq)))
            self.Errores.append(error)
            
            if imprimir_error_cada!=0:
                if (print_flag<imprimir_error_cada):
                    print_flag+=1
                else:
                    print(error)
                    print_flag=1

            de_db1=(1/(np.size(y_hat)))*np.dot(((y_hat-self.Training_Data_Y)*self.Training_Data_X),np.ones(np.size(y_diff)))
            de_db0=(1/(np.size(y_hat)))*np.dot((y_hat-self.Training_Data_Y),np.ones(np.size(y_diff)))

            beta_0=beta_0-self.Alpha*de_db0
            beta_1=beta_1-self.Alpha*de_db1
            beta=np.array([beta_1,beta_0])
            self.Betas.update({i:beta})

            y_hat=np.dot(np.transpose(x_hat),beta)
        
        return self.Errores, self.Betas
    #==============================================================================================

    # PLOT ERROR ==================================================================================
    def ShowErrorPlot(self,error,plotTittle,strColor):
        plt.plot(error,color=strColor)
        plt.title(plotTittle)
        plt.yscale('linear')
        plt.xlabel('No. Iteracion')
        plt.ylabel('Error')
        plt.show()
    # =============================================================================================

    # FUNCION PLOTEAR MODELO POR ITERACION ========================================================
    def ModeloPlot(self,n):
        x_range=np.linspace(np.min(self.Training_Data_X),np.max(self.Training_Data_X),100)
        it_no=[]

        for i in range(0,int(len(self.Betas)/n)):
            plt.plot(x_range, Recta(x_range,self.Betas[((i+1)*n)-1][0],self.Betas[((i+1)*n)-1][1]))
            it_no.append((i+1)*n)
        plt.legend(it_no)
        plt.show()
    # =============================================================================================

    # FUNCION ENTRENAMIENTO SKLEARN ===============================================================
    def training_LM_sklear(self):
        x = self.Training_Data_X.reshape(-1,1)
        y = self.Training_Data_Y
        reg = LinearRegression().fit(x,y)

        return reg
    # =============================================================================================

    # COMPARAR MODELOS ============================================================================
    def CompareModels(self,ModeloManual,ModeloSK,x_predict):
    
        x_pred_hat=[x_predict,np.ones(np.size(x_predict))]

        error, betas = ModeloManual(0)
        reg = ModeloSK()

        beta=np.array([betas[len(betas)-1][0],betas[len(betas)-1][1]])

        y_manual=np.dot(np.transpose(x_pred_hat),beta)
        y_SK=reg.predict(x_predict.reshape(-1,1))
        y_avrg=(y_manual+y_SK)/2

        return y_manual, y_SK, y_avrg
    # =============================================================================================

    # COMPARAR ERROR DE CADA MODELO ===============================================================
    def ErrorModelos(self,x_predict,y_real):
        y1,y2,y3 = self.CompareModels(self.training_LM,self.training_LM_sklear,x_predict)
        e_list_manual = []
        e_list_SK = []
        e_list_avrg =[]
        for i in range(0,len(y1)):
            e_list_manual.append(0.5*(y_real - y1[i])**2)
            e_list_SK.append(0.5*(y_real - y2[i])**2)
            e_list_avrg.append(0.5*(y_real - y3[i])**2)

        self.ShowErrorPlot(error_list_manual,'Error Modelo Manual','blue')
        self.ShowErrorPlot(error_list_SK,'Error Modelo SKLearn','red')
        self.ShowErrorPlot(error_list_avrg,'Error Modelo Promedio','green')

        e_Manual = (1/(2*np.size(y1)))*np.dot(((y_real-y1)**2),np.ones(np.size(y1)))
        e_SK = (1/(2*np.size(y2)))*np.dot(((y_real-y2)**2),np.ones(np.size(y2)))
        e_avrg = (1/(2*np.size(y3)))*np.dot(((y_real-y3)**2),np.ones(np.size(y3)))

        print(f'el error total manual es de {e_Manual}')
        print(f'el error total SK es de {e_SK}')
        print(f'el error total promedio es de {e_avrg}')
    # =============================================================================================
```
