## Importacion y separacion de la data

```python
import numpy as np

data = np.load('tmp/proyecto_training_data.npy')
rango = data.ndim
dimensiones = data.shape
print(f'el rango del tensor es: {rango} y las dimensiones son: {dimensiones}')
```

```python
# separando la data de training y la data de  prubas
Training_data=data[0:1168,0:7]
rango = Training_data.ndim
dimensiones = Training_data.shape
print(f'el rango del tensor es: {rango} y las dimensiones son: {dimensiones}')

Test_data=data[1168:1460,0:7]
rango = Test_data.ndim
dimensiones = Test_data.shape
print(f'el rango del tensor es: {rango} y las dimensiones son: {dimensiones}')
```
